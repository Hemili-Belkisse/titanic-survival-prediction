Task 1:  **Load and explore the data**
Import the Pandas library
Load train.csv into a variable called train
Load test.csv into a variable called test
Display the first 5 rows of the training data
Print the shape of the dataset (how many rows and columns?)
Check which columns have missing values and how many

```python
import pandas as pd
train=pd.read_csv('/content/train.csv')
test=pd.read_csv('/content/test.csv')
print("the first train rows\n",train.head())
print(train.describe())
print(train.shape)
missing_values=train.isnull().sum()
print(missing_values)
```

Task 2: **Understand the data**
Answer these questions by looking at the data:

How many passengers are in the training set?
How many survived vs died?  0: died,  1: survived
What are the column names and what does each one mean?

```python
print("how many passengers:",len(train))
print("\n Survived vs Died ")
print(train['Survived'].value_counts())
```

Task 3: **Clean the data**

Fill missing values in the Age column with the median age (do this for both train and test)
Fill missing values in the Fare column in test only
Convert the Sex column from text (male/female) to numbers (0/1)

```python
from numpy import median
test['Age']=test['Age'].fillna(test['Age'].median())

train['Age']=train['Age'].fillna(train['Age'].median())

test['Fare']=test['Fare'].fillna(test['Fare'].median())

test['Sex'] = test['Sex'].map({'male': 0, 'female': 1})
train['Sex']=train['Sex'].map({'male':0,'female':1})
```

Task 4: **Prepare features**

Choose these 4 features to train on: Age, Sex, Pclass, Fare
Create variable X with those features from the train set
Create variable y with the Survived column
Split X and y into training and validation sets (80% train, 20% validation)

```python
features=['Age', 'Sex', 'Pclass', 'Fare']
x= train[features]
y=train['Survived']

from sklearn.model_selection import train_test_split

x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=1)

print("\n=== Split sizes ===")
print("Training samples  :", len(x_train))
print("Validation samples:", len(x_val))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model=LogisticRegression()
model.fit(x_train,y_train)
pred=model.predict(x_val)
accuracy=accuracy_score(y_val,pred)
print("\n=== Model Accuracy ===")
print(f"accuracy: {accuracy*100:.1f}%")
```

```python

X_test           = test[features]
test_predictions = model.predict(X_test)

submission = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Survived'   : test_predictions
})
submission.to_csv('submission.csv', index=False)

print("\n=== Submission file ===")
print(submission.head(10))
print("submission.csv saved! Upload it to Kaggle.")
```

```python
features=['Age', 'Sex', 'Pclass', 'Fare']
x= train[features]
y=train['Survived']

from sklearn.model_selection import train_test_split

x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=1)

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
model=DecisionTreeClassifier(random_state=1)
model.fit(x_train,y_train)
pred=model.predict(x_val)
tr_accuracy=accuracy_score(y_val,pred)
print("\n=== Comparison ===")
print(f"Logistic Regression accuracy : {accuracy * 100:.1f}%")
print(f"Decision Tree accuracy       : {tr_accuracy * 100:.1f}%")
```
